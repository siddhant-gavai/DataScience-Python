# 13 - Set and Set Methods

Sets are unordered collections of unique elements. They are mutable but the elements themselves must be immutable.

### Creating Sets

```python
# Creating an empty set (Note: {} creates an empty dictionary)
s1 = set()
print(f"Empty set: {s1}, Type: {type(s1)}")

# Creating a set with elements
s2 = {1, 2, 3, 3, 2, 1} # Duplicates are automatically removed
print(f"Set with elements: {s2}")

# Creating a set from a list
my_list = [10, 20, 30, 20, 10]
s3 = set(my_list)
print(f"Set from list {my_list}: {s3}")
```

### Common Set Methods

```python
s = {1, 2, 3}
print(f"Original set: {s}")

# 1. add(): Adds an element to the set
s.add(4)
print(f"After add(4): {s}")

# 2. update(): Adds multiple elements
s.update([5, 6, 4])
print(f"After update([5, 6, 4]): {s}")

# 3. remove(): Removes an element (Raises KeyError if not found)
s.remove(2)
print(f"After remove(2): {s}")

# 4. discard(): Removes an element (No error if not found)
s.discard(10)
print(f"After discard(10): {s}")

# 5. pop(): Removes and returns an arbitrary element
popped = s.pop()
print(f"Popped element: {popped}, Remaining set: {s}")

# 6. clear(): Removes all elements
s.clear()
print(f"After clear(): {s}")
```

### Set Mathematical Operations

```python
setA = {1, 2, 3, 4, 5}
setB = {4, 5, 6, 7, 8}
print(f"Set A: {setA}")
print(f"Set B: {setB}")

# Union (|): All elements from both sets
print(f"Union (A | B): {setA | setB}")

# Intersection (&): Common elements
print(f"Intersection (A & B): {setA & setB}")

# Difference (-): Elements in A but not in B
print(f"Difference (A - B): {setA - setB}")

# Symmetric Difference (^): Elements in either A or B, but not both
print(f"Symmetric Difference (A ^ B): {setA ^ setB}")
```
