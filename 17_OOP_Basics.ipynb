# 17 - Object-Oriented Programming (OOP)

OOP is a programming paradigm based on the concept of "objects", which can contain data and code. Python is an object-oriented language.

### 1. Classes and Objects
A **Class** is like a blueprint, and an **Object** is an instance of that blueprint.

```python
class Dog:
    pass

# Creating an object
my_dog = Dog()
print(type(my_dog))
```

### 2. The `__init__` Method (Constructor)
It is used to initialize the object's state. It runs automatically when a new object is created.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name   # Instance variable
        self.breed = breed # Instance variable

dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Rocky", "German Shepherd")

print(f"{dog1.name} is a {dog1.breed}")
print(f"{dog2.name} is a {dog2.breed}")
```

### 3. Instance Methods
Functions defined inside a class that operate on the instance.

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f"{self.name} says Woof!"

my_dog = Dog("Buddy")
print(my_dog.bark())
```

### 4. Inheritance
A class can inherit attributes and methods from another class.

```python
# Parent Class
class Animal:
    def speak(self):
        return "Animal speaks"

# Child Class
class Cat(Animal):
    
    def speak(self):
        return "Meow!"

my_cat = Cat()
print(my_cat.speak()) # Overridden method
```

### 5. Encapsulation (Private Members)
Using double underscores `__` to make variables private.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance # Private variable

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
print(f"Balance: {account.get_balance()}")
# print(account.__balance) # This would throw an error
```
