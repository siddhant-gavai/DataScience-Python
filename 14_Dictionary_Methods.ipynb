# 14 - Dictionary and Dictionary Methods

Dictionaries are used to store data values in `key:value` pairs. A dictionary is a collection which is ordered (as of Python 3.7), changeable and does not allow duplicates.

### Creating Dictionaries

```python
# Creating an empty dictionary
dict1 = {}
print(f"Empty dictionary: {dict1}, Type: {type(dict1)}")

# Creating a dictionary with values
person = {
    "name": "Siddhant",
    "age": 22,
    "city": "Mumbai"
}
print(f"Dictionary person: {person}")

# Using the dict() constructor
car = dict(brand="Ford", model="Mustang", year=1964)
print(f"Dictionary car: {car}")
```

### Accessing and Modifying Items

```python
person = {"name": "Siddhant", "age": 22, "city": "Mumbai"}

# Accessing values
print(f"Name: {person['name']}")
print(f"Age using get(): {person.get('age')}")
print(f"Unknown key using get(): {person.get('profession', 'Not Found')}")

# Modifying values
person["age"] = 23

# Adding new items
person["profession"] = "Software Engineer"
print(f"Modified dictionary: {person}")
```

### Common Dictionary Methods

```python
info = {"id": 101, "dept": "CS", "rank": 5}
print(f"Original dictionary: {info}")

# 1. keys(): Returns a list of all keys
print(f"1. Keys: {info.keys()}")

# 2. values(): Returns a list of all values
print(f"2. Values: {info.values()}")

# 3. items(): Returns a list of key-value pairs (tuples)
print(f"3. Items: {info.items()}")

# 4. update(): Updates the dictionary with the specified items
info.update({"rank": 1, "year": 2024})
print(f"4. After update(): {info}")

# 5. pop(): Removes the element with the specified key
info.pop("dept")
print(f"5. After pop('dept'): {info}")

# 6. popitem(): Removes the last inserted item
info.popitem()
print(f"6. After popitem(): {info}")

# 7. clear(): Removes all elements
info.clear()
print(f"7. After clear(): {info}")
```
