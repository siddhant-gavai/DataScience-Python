# 16 - JSON Handling in Python

JSON (JavaScript Object Notation) is a lightweight data-interchange format. Python has a built-in package called `json`, which can be used to work with JSON data.

### Importing the json module

```python
import json
```

### Reading JSON from a File

We use `json.load()` to read data from a file object.

```python
with open('data.json', 'r') as f:
    data = json.load(f)

print("Data loaded from data.json:")
print(data)
print(f"\nName: {data['name']}")
print(f"Skills: {', '.join(data['skills'])}")
```

### Writing JSON to a File

We use `json.dump()` to write data to a file object.

```python
new_data = {
    "name": "Siddhant Gavai",
    "role": "Data Scientist",
    "updated": True
}

with open('updated_data.json', 'w') as f:
    json.dump(new_data, f, indent=4)

print("New data written to updated_data.json")
```

### Parsing JSON Strings

- `json.loads()`: Parses a JSON string.
- `json.dumps()`: Converts a Python object into a JSON string.

```python
# JSON string to Python dictionary
json_string = '{"city": "New York", "population": 8419000}'
python_dict = json.loads(json_string)
print(f"City: {python_dict['city']}")

# Python dictionary to JSON string
back_to_json = json.dumps(python_dict, indent=2)
print("\nJSON String:")
print(back_to_json)
```
