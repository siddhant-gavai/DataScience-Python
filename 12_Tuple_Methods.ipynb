## Tuple Packing and Unpacking

- **Packing**: When we create a tuple, we normally assign values to it. This is called "packing" a tuple.
- **Unpacking**: We can also extract the values back into variables. This is called "unpacking".

```python
# Tuple Packing
fruits = ("apple", "banana", "cherry")
print(f"Packed tuple: {fruits}")

# Tuple Unpacking
(green, yellow, red) = fruits

print(f"Unpacked values:")
print(f"green: {green}")
print(f"yellow: {yellow}")
print(f"red: {red}")
```

### Extended Unpacking (using `*`)

If the number of variables is less than the number of values, you can add an `*` to the variable name and the values will be assigned to the variable as a list.

```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(f"green: {green}")
print(f"yellow: {yellow}")
print(f"red (list): {red}") # Remaining items collected as a list
```
