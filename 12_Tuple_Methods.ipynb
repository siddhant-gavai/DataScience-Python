# 12 - Tuple and Tuple Methods

This notebook covers tuples (immutable sequences) and their methods.

```python
coordinates = (10, 20, 10, 30)

print(f"Count of 10: {coordinates.count(10)}")
print(f"Index of 20: {coordinates.index(20)}")

# Note: Tuples are immutable, so you cannot append or change them.
```
