# 15 - File Handling in Python

File handling is an important part of any web application. Python has several functions for creating, reading, updating, and deleting files.

### File Opening Modes

The `open()` function takes two parameters: **filename** and **mode**.

- `"r"` - **Read** - Default value. Opens a file for reading, error if the file does not exist.
- `"a"` - **Append** - Opens a file for appending, creates the file if it does not exist.
- `"w"` - **Write** - Opens a file for writing, creates the file if it does not exist.
- `"x"` - **Create** - Creates the specified file, returns an error if the file exists.
- `"t"` - **Text** - Default value. Text mode.
- `"b"` - **Binary** - Binary mode (e.g. images).

### Writing to a File

```python
# Using 'with' statement ensures the file is properly closed after usage
with open("demo.txt", "w") as f:
    f.write("Hello! This is a demo file created using Python.\n")
    f.write("File handling is simple and powerful.")

print("File 'demo.txt' created and written successfully.")
```

### Reading from a File

```python
# Reading the entire file
with open("demo.txt", "r") as f:
    content = f.read()
    print("--- Full Content ---")
    print(content)

# Reading line by linex
with open("demo.txt", "r") as f:
    print("\n--- Line by Line ---")
    for line in f:
        print(line.strip())
```

### Appending to a File

```python
with open("demo.txt", "a") as f:
    f.write("\nAdding a new line to the existing file.")

with open("demo.txt", "r") as f:
    print("--- Updated Content ---")
    print(f.read())
```

### Deleting a File

```python
import os

file_to_delete = "demo.txt"

if os.path.exists(file_to_delete):
    # os.remove(file_to_delete) # Uncomment to actually delete
    print(f"File '{file_to_delete}' exists and can be deleted using os.remove()")
else:
    print("The file does not exist")
```
