# 09 - String Formatting and f-strings

This notebook covers different ways to format strings in Python.

```python
name = "Siddhant"
age = 22

# f-strings (Modern way)
print(f"My name is {name} and I am {age} years old.")

# .format() method
print("My name is {} and I am {} years old.".format(name, age))

# % operator (Old way)
print("My name is %s and I am %d years old." % (name, age))
```

## Padding and Alignment

You can use f-strings to align text and add padding.

- `<` : Left aligned within the given width (default for strings)
- `>` : Right aligned within the given width (default for numbers)
- `^` : Center aligned within the given width

```python
text = "Python"

print(f"|{text:<10}|")   # Left aligned in 10 spaces
print(f"|{text:>10}|")   # Right aligned in 10 spaces
print(f"|{text:^10}|")   # Center aligned in 10 spaces

# Using custom padding character
print(f"{text:*^20}")   # Center aligned with * padding
print(f"{text:=<20}")   # Left aligned with = padding

# Number padding (adding leading zeros)
num = 7
print(f"{num:03}")       # Output: 007
```
