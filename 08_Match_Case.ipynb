# 08 - Match Case (Python 3.10+)

This notebook explores structural pattern matching using `match` and `case`.

```python
status = 404

match status:
    case 200:
        print("Success")
    case 404:
        print("Not Found")
    case 500:
        print("Server Error")
    case _:
        print("Unknown Status")
```
