# 07 - Functions in Python

This notebook covers how to define and use functions.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Siddhant"))
```

```python
def add(a, b):
    return a + b

result = add(10, 20)
print(f"Sum: {result}")
```
