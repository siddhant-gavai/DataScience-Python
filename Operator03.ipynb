```python
# Arithmatic operator....
a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.33
print(a % b)   # 1 (remainder)
print(a ** b)  # 1000 (power)
print(a // b)  # 3 (floor division)
```

```python
# Comparison operator 
a = 10
b = 5

print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False
```

```python
#Logical Operator 
x = 10

print(x > 5 and x < 20)  # True || Both are true 
print(x > 5 or x > 20)   # True || one true is satisfied 
print(not(x > 5))        # False || Reverse
```

```python
# 4. Assignment Operators (value assign/update)
x = 5

x += 3   # x = x + 3 → 8
x -= 2   # 6
x *= 2   # 12
x /= 3   # 4.0
```

```python
# 6. Identity Operators (is, is not)
a = [1,2]
b = [1,2]

print(a == b)  # True (value same)
print(a is b)  # False (memory different)
```
