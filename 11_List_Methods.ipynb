# 11 - List and List Methods

This notebook covers list creation and common methods.

```python
fruits = ["apple", "banana", "cherry"]

fruits.append("orange")  # Add to end
print(f"After append: {fruits}")

fruits.pop()             # Remove last
print(f"After pop: {fruits}")

fruits.insert(1, "blueberry") # Insert at index
print(f"After insert: {fruits}")
```

### More Common List Methods

Here are several more methods commonly used with lists in Python.

```python
numbers = [5, 2, 8, 2, 1]
print(f"Original list: {numbers}")

# 1. extend(): Add all elements of an iterable to the end
numbers.extend([10, 20])
print(f"1. After extend([10, 20]): {numbers}")

# 2. remove(): Removes the first item with the specified value
numbers.remove(2)
print(f"2. After remove(2): {numbers}")

# 3. index(): Returns the index of the first element with the specified value
idx = numbers.index(8)
print(f"3. Index of 8: {idx}")

# 4. count(): Returns the number of elements with the specified value
cnt = numbers.count(2)
print(f"4. Count of 2: {cnt}")

# 5. sort(): Sorts the list in ascending order (in-place)
numbers.sort()
print(f"5. After sort(): {numbers}")

# 6. reverse(): Reverses the order of the list (in-place)
numbers.reverse()
print(f"6. After reverse(): {numbers}")

# 7. copy(): Returns a shallow copy of the list
num_copy = numbers.copy()
print(f"7. Copy of list: {num_copy}")

# 8. clear(): Removes all elements from the list
num_copy.clear()
print(f"8. After clear() on copy: {num_copy}")
```
